```python
#Import SQLLite module
import sqlite3
```

```python
#Connect to sqlite database

conn = sqlite3.connect('melbournebusinessdatabase.sqlite')
c = conn.cursor()
```

```python
#Print unique trading_name

c.execute("SELECT DISTINCT trading_name FROM melbourne_business")

rows = c.fetchall()

for row in rows:
    print(row)
```

```python
#Print count

c.execute("SELECT COUNT(*) FROM (SELECT DISTINCT trading_name FROM melbourne_business)")

rows = c.fetchall()

print(rows)
```
